```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_moons
```

```python
x,y = make_moons(n_samples= 250,noise=0.05)  # where x is the data and y is target
dataset = {"data1":x[:,0] , "data2":x[:,1]}
```

```python
df = pd.DataFrame(dataset)
```

```python
df.head(3)
```

```python
from sklearn.cluster import DBSCAN
```

```python
Db = DBSCAN(eps=0.2,min_samples=5)
df["predict"] = Db.fit_predict(df)
df.head(3)
```

```python
sns.scatterplot(x="data1",y="data2",data=df,hue="predict")
```
