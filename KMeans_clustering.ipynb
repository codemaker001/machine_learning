```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("iris.csv")
```

```python
df = df.drop(columns = "Species")
df.head(3)
```

# KMEANS

  used for linearly seperable data , we can see that by making a pairplot graph and seeing by ourselfs

```python
from sklearn.cluster import KMeans
```

TO FIND THE NUMBER OF K CLUSTERS

```python
wcss = []

for i in range(2,21):
    kmc = KMeans(n_clusters=i)
    kmc.fit(df)
    wcss.append(kmc.inertia_)
```

```python
plt.plot([i for i in range(2,21)],wcss,marker="o")
plt.xticks([i for i in range(2,21)])
plt.grid()
plt.show
```

WE CAN SEE THAT 3 IS THE ELBOW POINT HENCE IDEAL NUMBER OF CLUSTER SHOULD BE 3

```python
km = KMeans(n_clusters=3)
km.fit(df)
```

```python
df["predict"] = km.predict(df)
```

```python
sns.pairplot(df,hue="predict")
```
